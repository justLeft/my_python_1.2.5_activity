```python
import pandas as pd
import numpy as np

movies = pd.read_excel('movies_merge.xlsx')
ott = pd.read_csv('ott_merge.csv')

print(movies.columns)
print(movies.shape)
print(ott.columns)
print(ott.shape)

mov_ott = pd.merge(movies, ott, how='left', on = 'ID')

print(mov_ott.columns)
print(mov_ott.shape)
```

```python
mov_ott
```

```python
mov_ott_runtime = mov_ott[['ID', 'Title', 'Runtime', 'Genres']]    # Define df to work with, adding Title also

# View the output.
mov_ott_runtime
```

```python
# question 1, Add one minute to the runtime for ads

mov_ott_runtime['Runtime'].add(1)    # Add 60 seconds or 1 minute to runtime, use add() because its series.
```

```python
# question 2, Which movies are documentaries?

# create new column in df Gen_doc with documentaries or not.
# specify the column py needs to look, the value, the value to return if true, and the value if false
mov_ott_runtime['Gen_doc'] = np.where(mov_ott_runtime['Genres'].str.contains('Documentary'),
                                      'Documentary', 'Not Documentary')

# View the DataFrame.
mov_ott_runtime
```
