```python
import pandas as pd
import numpy as np

movies = pd.read_excel('movies_merge.xlsx')
ott = pd.read_csv('ott_merge.csv')

mov_ott = pd.merge(movies, ott, how='left', on = 'ID')
```

```python
# View IMDb and Rotten Tomatoes columns.
mov_ott_ratings = mov_ott[['ID', 'IMDb', 'Rotten Tomatoes']]

# View the DataFrame.
mov_ott_ratings
```

```python
# Replace missing values with 0.
mov_ott_ratings_final = mov_ott_ratings.fillna(0)

# View the DataFrame.
mov_ott_ratings_final
```

```python
# question 1: What is the average rating per movie?

# define a function to calculate average rating
# convert IMDb to decimal (/10), add Rotten Tomaties) and /2
# Add a new column to the DataFrame indicating average rating

def avg_rating(imdb_score,RT_score):
    avg_rat = (imdb_score/10 + RT_score)/2
    return avg_rat

mov_ott_ratings_final['Rating'] = avg_rating(mov_ott_ratings_final['IMDb'], mov_ott_ratings_final['Rotten Tomatoes'])

# View the DataFrame.
mov_ott_ratings_final                    
```

```python
# question 2: How many movies were released per content rating (recommended Age to watch)?

# Categorical count. 
def count_movies(df1):
    print(df1.value_counts())

# Number of movies released per 'Age'.
df = mov_ott['Age'].astype('category')

# View the output.
count_movies(df)
```

```python
# How many movies were released per year?

# Number of movies released per 'Year'.
yr = mov_ott['Year'].astype('category')

# View the output.
count_movies(yr)
```
